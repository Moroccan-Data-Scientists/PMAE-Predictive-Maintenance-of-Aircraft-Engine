## Data Visualization


```python
# visualization matplotlib subplots
df_One.std().plot(kind='bar', figsize=(10,6), title="Features Standard Deviation")
```

```python
table= ['s12',
 's7',
 's21',
 's20',
 's6',
 's14',
 's9',
 's13',
 's8',
 's3',
 's17',
 's2',
 's15',
 's4',
 's11']

for col in df_One.columns: 
    table.append(col)
```

```python
#visualization subset of features

Size = np.corrcoef(df_One[table].values.T)
sns.set(font_scale=1.0)
fig = plt.figure(figsize=(17, 15))
hm = sns.heatmap(Size, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 8}, yticklabels=table, xticklabels=table)
plt.title('Features Correlation Heatmap')
plt.show()
```

```python
df_Two.std().plot(kind='bar', figsize=(10,6), title="Features Standard Deviation")
```

```python
tableTwo= ['s12',
 's7',
 's21',
 's20',
 's6',
 's14',
 's9',
 's13',
 's8',
 's3',
 's17',
 's2',
 's15',
 's4',
 's11']

for colu in df_Two.columns: 
    tableTwo.append(colu)
```

```python
SizeT = np.corrcoef(df_Two[tableTwo].values.T)
sns.set(font_scale=1.0)
fig = plt.figure(figsize=(10, 8))
hm = sns.heatmap(SizeT, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 8}, yticklabels=tableTwo, xticklabels=tableTwo)
plt.title('Features Correlation Heatmap')
plt.show()
```
